```python
import numpy as np
import matplotlib.pyplot as plt
import math
# Étape 1 : Créer une grille
zoom = 10

F = (0,0) # foyer
d = {"a":1,"b":0,"c":5} # droite directrice (a,b,c) tq ax+by+c=0 
e = (6,6.25,7)

def evaluate(x,y):
    """
    Return e = d(P,F)/d(P,d)
    Retourne l'excentricité pour lequel le point est sur la conique
    """
    dpf = np.sqrt((x-F[0])**2+(y-F[1])**2)
    dpd = abs(d["a"]*x+d["b"]*y+d["c"]) / np.sqrt(d["a"]**2 + d["b"]**2)
    return dpf*dpd

def droite_directrice(x,y):
    return d["a"]*x+d["b"]*y+d["c"]

x = np.linspace(-zoom, zoom, 400)
y = np.linspace(-zoom, zoom, 400)
X, Y = np.meshgrid(x, y)
Z = evaluate(X,Y)
Zd = droite_directrice(X,Y)
plt.plot(F[0], F[1], 'go', label='Foyer F')
plt.text(F[0]+0.2, F[1], 'F', color='green')

plt.contour(X, Y, Z, levels=e, colors=['blue',"green"]) # conique
plt.contour(X, Y, Zd, levels=[0], colors='red') # droite directrice


plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.gca().set_aspect('equal')
plt.xlabel('x')
plt.ylabel('y')
# Affichage de la grille correspondant à np.linspace
# for i in range(len(x)):
#     plt.axvline(x[i], color='gray', linestyle='--', linewidth=0.5)
#     plt.axhline(y[i], color='gray', linestyle='--', linewidth=0.5)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Création de la figure et de l'axe
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

# Données initiales
x = np.linspace(0, 2 * np.pi, 400)
freq_init = 1
y = np.sin(freq_init * x)
line, = ax.plot(x, y)

# Position et paramètres du curseur
ax_slider = plt.axes([0.25, 0.1, 0.65, 0.03])
slider = Slider(
    ax=ax_slider,
    label='Fréquence',
    valmin=0,
    valmax=2,
    valinit=freq_init,
    valstep=0.01,
)

# Fonction de mise à jour appelée à chaque changement de valeur du curseur
def update(val):
    freq = slider.val
    line.set_ydata(np.sin(freq * x))
    fig.canvas.draw_idle()

# Lier la fonction de mise à jour au curseur
slider.on_changed(update)

# Afficher
plt.show()

```
